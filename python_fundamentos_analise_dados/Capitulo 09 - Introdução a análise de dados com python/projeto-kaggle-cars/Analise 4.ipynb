```python
# Imports
import os
import subprocess
import stat
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
sns.set(style="white")
%matplotlib inline
```

```python
# Dataset
clean_data_path = "dataset/autos.csv"
df = pd.read_csv(clean_data_path,encoding="latin-1")
df.head()
```

## Preço médio de um veículo por marca, bem como tipo de veículo

```python
flights = sns.load_dataset("flights")
print(flights.head())
flights = flights.pivot("month", "year", "passengers")
ax = sns.heatmap(flights)
```

```python
# Crie um Heatmap com Preço médio de um veículo por marca, bem como tipo de veículo

df2 = df.groupby(['vehicleType', 'brand']).mean().reset_index()
df2.head()
prices = df2[df2['price'] < 10000].pivot('brand', 'vehicleType', 'price')

fig, ax = plt.subplots(figsize=(20,40)) 
ax1 = sns.heatmap(prices, ax=ax)
fig = ax1.get_figure()
```

```python
# Salvando o plot
fig.savefig("plots/Analise4/heatmap-price-brand-vehicleType.png")
```
